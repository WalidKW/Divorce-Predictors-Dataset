### 3. Consider splitting the data into Training, Validation, and Testing sets. The suggested split is 70%, 10%, and 20% held-out testing set, respectively.

```python
X = df.iloc[:,:-1].values
y = df['Class'].values

# Splitting the datasetl into the Training set and Test set for model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
# validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.1, random_state = 100)
```

### <span style='color:yellow'> Task#: Logistic Regression according to your formulation in step 1</span> 

```python
# Create the logistic regression model
clf = LogisticRegression()

# Fit the model on the training data
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)
# Calculate the accuracy, f1-score, recall, and precision of the model
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score

print("Accuracy:", accuracy)
print("F1 score:", f1)
print("Recall:", recall)
print("Precision:", precision)

plot_confusion_matrix(clf, X_test, y_test, display_labels=["class1", "class2"])
```
